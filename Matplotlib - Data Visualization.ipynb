```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

### Two methods to plot:
    One is a functional method and another is object oriented method.

```python
#Functional Method:
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show() # for jupyter notebook not mandatory
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
plt.show()
```

```python
# Object Oriented
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.5,0.3])

axes1.plot(x,y,'r')
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
# Creating subplot
fig, axes = plt.subplots(nrows=1,ncols=2)
# axes.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout() # to fix overlappings
```

## Figure size and DPI

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('matplot_fig1.png',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
# ax.set_title('Title')
# ax.set_ylabel('Y')
# ax.set_xlabel('X')
ax.legend()
```

## Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=0.5, ls='-', marker='o', markersize=10,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2, ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
